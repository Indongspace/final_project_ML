## 데이터 확인

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data/for_statistic/연령별 경제활동인구 20대.csv', encoding='euc-kr')
df.head(3)
```

```python
df.set_index('항목', inplace=True)
df
```

```python
# 열의 값들을 1명 단위로 변경 (곱하기1000)
def multiply_1000(df, column_name):
    df[column_name] = df[column_name]*1000
    return df
```

```python
df = multiply_1000(df, '경제활동인구')
df = multiply_1000(df, '취업자')
df = multiply_1000(df, '실업자')
df = multiply_1000(df, '비경제활동인구')
df.head()
```
